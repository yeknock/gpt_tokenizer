```python
# changing text into 
tokens = text.encode("utf-8")
tokens = list(map(int, tokens))
print(tokens)
```

```python
def get_stats(tokens):
    pair_dict = {}
    counts = {}

    # Creating a dict of each pair in tokens
    for i in range(len(tokens) - 1):
        pair_dict[i] = (tokens[i], tokens[i + 1])

    # create a dict of occurance of each pair
    for value in pair_dict.values():
        counts[value] = counts.get(value, 0) + 1

    return counts
```

```python
stats = get_stats(tokens)
print(stats)
```
